# Simple Tokenizer
To create a simple tokenizer, we can use regular expressions.

Regular expression (shortened to regex) is a tool for describing the shape of text we want to extract. Instead of checking characters one by one, we can tell Python what text to find.

First, we need to import Python's built-in regular expression re module:

import re

For creating tokens, we want either a word, or a piece of punctuation. The regex for that is:

\w+|[^\w\s]

\w+: one or more word characters
|: or
[^\w\s]: a single character that is neither a word character nor whitespace
Simply put, look for a word or a punctuation mark.

Then, we can use re.findall() to go through a text and return the snippets found in a list:

text = 'The group chat is unhinged.'

re.findall(r'\w+|[^\w\s]', text)

If we print out the last line, it would look like:

['The', 'group', 'chat', 'is', 'unhinged', '.']



```python
import re

def tokenize(text):
  return re.findall(r'\w+|[^\w\s]',text)
print(tokenize('My name is Nandana R Nair.'))
```
